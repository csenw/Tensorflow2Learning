# Basics of Tensorflow 2.0

```python
# 1. Creat a constant/scalar, vector, matrix
a = tf.constant(1.2)
aa = tf.constant ([2,3])
aaa = tf.constant([[1,2],[3,4]])
print(a,aa,aaa)
# String will be used as a constant of string
b = tf.constant('This is a string!')
print(b)
# tf.strings provides some 
tf.strings.lower(b)
# Boolean type
c = tf.constant(True)
cc = tf.constant(False)
print(c,cc)
```

```python
# 2. Precision
import numpy as np
a = tf.constant(123,dtype=tf.int16)
b = tf.constant(123,dtype=tf.int32)
c = tf.constant(np.pi,dtype=tf.float32)
d = tf.constant(np.pi,dtype=tf.float64)
# to read the precision of a variable/tensor, use xxx.dtype (xxx is the name of the tensor)
print("data type: ",a.dtype)
# use tf.cast to convert the data type: new_tensor = tf.cast(xxx, new_type)
if a.dtype!=tf.float32:
    a = tf.cast(a, tf.float32)
print("data type: ",a.dtype)
```

```python
# 3. Create a tensor
# 3.1 Create a tensor from an existing data object
tf.convert_to_tensor([1.0, 2.0], dtype=tf.float16)
tf.convert_to_tensor(np.array([3,4]), dtype=tf.float32)
# 3.2 Create all ones/zeros tensor
a = tf.ones([2,2])
b = tf.zeros([3,3])
# use tf.zeros_like/ones_like to create all ones/zeros tensor with the same shape as the input
aa = tf.zeros_like(a)
bb = tf.ones_like(b)
print(a,aa)
print(b,bb)
# use tf.fill(shape, value) to create a tensor with all pre-defined values
c = tf.fill([4,4],4)
print(c)
# 3.3 Create a tensor with distributions.
# Use tf.random.normal(shape, mean, stddev) to create a tensor with a normal distribution
d = tf.random.normal([5,6], mean=1.0, stddev=1.0)
print(d)
e = tf.random.uniform([6,6], minval=10, maxval=20, dtype=tf.float32)
print(e)
# 3.4 Create a list with tf.range(limit, delta/step), begin with 0 by default
f = tf.range(10)
ff = tf.range(10,delta=2)
fff = tf.range(-10,10, 2)
print(f,ff,fff)
# 3.5 Indexing and Slicing
x = tf.random.normal([4,28,28,3])
print(x[0])
print(x[:2])
# 3.6 Reshape: tf.reshape(x, new_shape)
xx = tf.reshape(x,[4,-1,3])
xxx = tf.reshape(x,[-1,3])
print(xx,xxx)
# 3.7 Adding or Removing dimension
y = tf.random.normal([28,28])
print(y)
y = tf.expand_dims(y,-1)
print(y)
y = tf.expand_dims(y,0)
print(y)
# use tf.squeeze(x, axis) to delete a dimension
y = tf.squeeze(y,0)
print(y)
y = tf.squeeze(y,-1)
print(y)
# 3.8 Dimension swapping
x = tf.random.normal([2, 32, 28,2])
x = tf.transpose(x, [0,3,1,2])
print(x)
x = tf.transpose(x,[0,1,3,2])
print(x)
# 3.9 Clone data and Broadcasting
# use tf.tile(x, multiples) to clone data
y = tf.tile(y,[1,2])
print(y)
# as explicitly copying data is very expensive, it is more frequently to use broadcasting instead
# Broadcasting will align the smaller tensor in the right and add in more dimension as requested.
x = tf.random.normal([32,1])
# to convert x into a four dimension [1,32,32,1]
x = tf.broadcast_to(x,[1,32,32,1])
print(x)
```

# Advanced Operations in Tensorflow 2.0

```python
# 4. Merge and Divide
# 4.1 Use concatenate and stack to merge tensors
# tf.concate(tensors, axis) vs tf.stack(tensors, axis)
a = tf.random.normal([4,35,8])
b = tf.random.normal([6,35,8])
c = tf.concat([a,b], axis=0)
a = tf.random.normal([28,28])
b = tf.random.normal([28,28])
tf.stack([a,b], axis=-1) # tf.stack will produce a new dimension.
# 4.2 Use tf.split(x, num_or_size_splits, axis) to divide the tensor into smaller ones.
# num_or_size_splits can be a number or a list of numbers: 
# 10 - divide into 10 pieces; 
# [2, 4, 3, 1]- divide into four pieces, each of which have 2, 4, 3, and 1 items, respectively.
d = tf.split(c,10, axis=0)
d[1]
e = tf.split(c,num_or_size_splits=[2, 4, 3, 1], axis=0)
e[1]
# tf.unstack will be helpful to divide a tensor on one dimension into even pieces.
# Note: tf.unstack will automatically remove the dimension, while tf.split does not.
# Ex: tf.unstack([10, 4, 4], axis=0) -> 10 * [4, 4]
#     tf.split([10, 4, 4],10, axis=0)-> 10 * [1, 4, 4]
dd = tf.unstack(c, axis=0)
# print(dd)

# 4.3 Statistics
# norms: L1, L2, L-infinity
x = tf.random.normal([2,2])
#print(tf.norm(x, ord=1))
#print(tf.norm(x, ord=2))
#print(tf.norm(x, ord=np.inf))

# max/min, average/mean, sum
print(x)
print(tf.reduce_min(x, axis=0))
print(tf.reduce_min(x))
print(tf.reduce_max(x, axis=1))
print(tf.reduce_mean(x))
print(tf.reduce_sum(x))

# tf.argmax(x, axis) / tf.argmin(x, axis)
print(tf.argmax(x, axis=1))

# 4.4 Padding
x = tf.random.normal([3])
print(x)
px = tf.pad(x, [[1,2]])
print(px)

y = tf.random.normal([2,2])
print(y)
print(tf.pad(y,[[1,1],[1,2]]))

# 4.5 Use tf.maximum(x,a)/tf.minimum(x,a) to get the lower and upper bounds: get the maximum/minimum between x and a
x = tf.range(9)
print(x, tf.maximum(x,3), tf.minimum(x,6))

# 4.6 Use tf.gather(x, indices, axis)/tf.gather_nd to get the targeted data.
x = tf.random.normal([10,28,28,3])
# use tf.gather to collect first three images
print(tf.gather(x,[0,1,2],axis=0))
# use tf.gather_nd to collect data with multiple dimension specification
# use tf.gather_nd to collect the second row in first image
print(tf.gather_nd(x, [[0,1],[1,2]]))
# use tf.boolean_mask to indicate the indices
mask = [True, True, True, False, False, False, False, False, False, False]
print(tf.boolean_mask(x, mask, axis=0))
# use tf.where to read data from two tensors with specified conditions
# e.g. Football or Rest up to weather
a = tf.ones([3,3])
b = tf.zeros([3,3])
cond = tf.constant([[True, False, False],[False, True, False], [False, False, True]])
print(tf.where(cond, a,b))
# use tf.scatter_nd(indices, updates, shape) to update all zeros tensor
indices = tf.constant([[4],[3],[1],[7]])
updates = tf.constant([4.4, 3.3, 1.1, 7.7])
tf.scatter_nd(indices, updates,[8])
```
